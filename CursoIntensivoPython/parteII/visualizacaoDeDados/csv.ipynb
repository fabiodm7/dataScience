```python
import csv

filename = 'sitka_weather_07-2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    for index,column_header in enumerate(header_row):
        print(index,column_header)
```

```python
import matplotlib.pyplot as plt

with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    highs = []
    for row in reader:
        high = int(row[1])
        highs.append(high)

print(highs)

fig = plt.figure(figsize=(10,6))
plt.plot(highs,c='red')
plt.title('Temperaturas máximas em Julho de 2014',fontsize=24)
plt.xlabel('',fontsize=10)
plt.ylabel('Temperatura (*F)',fontsize=10)
plt.show
```

```python
from datetime import datetime

with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    dates,highs = [],[]
    for row in reader:
        current_date = datetime.strptime(row[0],'%Y-%m-%d')
        dates.append(current_date)

        high = int(row[1])
        highs.append(high)

print(highs)

fig = plt.figure(figsize=(10,6))
plt.plot(dates,highs,c='red')
plt.title('Temperaturas máximas em Julho de 2014',fontsize=24)
plt.xlabel('',fontsize=10)
fig.autofmt_xdate()
plt.ylabel('Temperatura (*F)',fontsize=10)
plt.show
```

```python
filename = 'sitka_weather_2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    dates,highs,lows = [],[],[]
    for row in reader:
        current_date = datetime.strptime(row[0],'%Y-%m-%d')
        dates.append(current_date)

        high = int(row[1])
        highs.append(high)

        low = int(row[3])
        lows.append(low)

fig = plt.figure(figsize=(10,6))
plt.plot(dates,highs,c='red',alpha=.9)
plt.plot(dates,lows,c='blue',alpha=.9)
plt.fill_between(dates,highs,lows,facecolor='green',alpha=.5)
plt.title('Temperaturas máximas em 2014 de sitka',fontsize=24)
plt.xlabel('',fontsize=10)
fig.autofmt_xdate()
plt.ylim(10,120)
plt.ylabel('Temperatura (*F)',fontsize=10)
plt.show
```

```python
filename = 'death_valley_2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    dates,highs,lows = [],[],[]
    for row in reader:
        try:
            current_date = datetime.strptime(row[0],'%Y-%m-%d')
            high = int(row[1])
            low = int(row[3])
        except ValueError:
            print(current_date,'missing data')
        else:
            dates.append(current_date)
            highs.append(high)
            lows.append(low)


fig = plt.figure(figsize=(10,6))
plt.plot(dates,highs,c='red',alpha=.9)
plt.plot(dates,lows,c='blue',alpha=.9)
plt.fill_between(dates,highs,lows,facecolor='green',alpha=.5)
plt.title('Temperaturas máximas em 2014 de Death Valley',fontsize=24)
plt.xlabel('',fontsize=10)
fig.autofmt_xdate()
plt.ylim(10,120)
plt.ylabel('Temperatura (*F)',fontsize=10)
plt.show
```
