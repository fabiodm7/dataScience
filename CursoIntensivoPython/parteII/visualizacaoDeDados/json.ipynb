```python
def get_country_code(country_name):
    """Devolve um código de duas letras do pygal"""
    for code,name in COUNTRIES.items():
        if name.lower() == country_name.lower():
            return code
    return None
```

```python
import json

filename = 'population_data.json'

with open(filename) as f:
    pop_data = json.load(f)
    
cc_populations = {}

for pop_dict in pop_data:
    if pop_dict['Year'] == '2010':
        country_name = pop_dict['Country Name']
        population = int(float(pop_dict['Value']))
        code = get_country_code(country_name)
        if code:
            cc_populations[code] = population
```

```python
# código feito por FDM

import json
from pygal.maps.world import COUNTRIES
from pygal.maps.world import World

filename = 'population_data.json'

with open(filename) as f:
    pop_data = json.load(f)

def get_nome_pais_pygal(pais):
    nome_pais = COUNTRIES[pais]
    return nome_pais

def get_codigo_pais_dados(pop_data,nome_pais,nome_pais_json,codigo_pais_json):
    for pop_dict in pop_data:
        if pop_dict[nome_pais_json] == nome_pais:
            return pop_dict[codigo_pais_json]

def get_populacao(pop_data,codigo_pais,ano,ano_json,codigo_pais_json,valor_json):
    for pop_dict in pop_data:
        if pop_dict[ano_json] == str(ano) and codigo_pais == pop_dict[codigo_pais_json]:
            population = int(float(pop_dict[valor_json]))
            return population

def cria_eixo_continental(populacao_continente,continente,pop_data,ano,nome_pais_json,codigo_pais_json,ano_json,valor_json):
    for pais in continente:
        nome_pais = get_nome_pais_pygal(pais)
        codigo_pais = get_codigo_pais_dados(
            pop_data,nome_pais,nome_pais_json,codigo_pais_json
        )
        populacao = get_populacao(
            pop_data,codigo_pais,ano,ano_json,codigo_pais_json,valor_json
        )
        populacao_continente[pais] = populacao

america_norte = ['ca','us','mx','gl']
america_central = [
    'bz','cr','gt','hn','hn','ni','sv','pa','cu','ht','jm','do','pr'
]
america_sul = [
    'ar','bo','br','cl','co','ec','gf','gy','pe','py','sr','uy','ve'
]

pais = COUNTRIES.keys()

ano = 2010
nome_pais_json = 'Country Name'
codigo_pais_json = 'Country Code'
ano_json = 'Year'
valor_json = 'Value'

populacao_america_norte = {}
populacao_america_central = {}
populacao_america_sul = {}
populacao_pais = {}

cria_eixo_continental(populacao_america_norte,america_norte,pop_data,ano,nome_pais_json,codigo_pais_json,ano_json,valor_json)

cria_eixo_continental(populacao_america_central,america_central,pop_data,ano,nome_pais_json,codigo_pais_json,ano_json,valor_json)

cria_eixo_continental(populacao_america_sul,america_sul,pop_data,ano,nome_pais_json,codigo_pais_json,ano_json,valor_json)

cria_eixo_continental(populacao_pais,pais,pop_data,ano,nome_pais_json,codigo_pais_json,ano_json,valor_json)

wm = World()
# wm.title = 'Populações da América do norte, central e do sul'
# wm.add('América do norte',populacao_america_norte)
# wm.add('América central',populacao_america_central)
# wm.add('América do sul',populacao_america_sul)
# wm.render_to_file('americas.svg')
wm.title = 'Populações'
wm.add('2010',populacao_pais)
wm.render_to_file('mundo.svg')
```
