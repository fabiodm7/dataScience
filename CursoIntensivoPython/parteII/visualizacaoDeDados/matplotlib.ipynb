# Passeios aleat√≥rios

```python
from random import choice

class RandomWalk():
    """Classe para gerar passeios aleat√≥rios"""
    def __init__(self,num_points=5000):
        """Inicializa os atributos de um passeio"""
        self.num_points = num_points

        # Todos os passeio come√ßam em (0,0)
        self.x_values = [0]
        self.y_values = [0]

    def fill_walk(self):
        """Calcula todos os pontos do passeio"""
        # Continua dando passos at√© que o passeio alcance o tamanho desejado
        while len(self.x_values) < self.num_points:
            # Decide dire√ß√£o a ser seguida e dist√¢ncia a ser percorrida nela
            x_direction = choice([1,-1])
            x_distance = choice([0,1,2,3,4])
            x_step = x_direction * x_distance

            y_direction = choice([1,-1])
            y_distance = choice([0,1,2,3,4])
            y_step = y_direction * y_distance

            # Rejeita movimentos que n√£o v√£o a lugar nenhum
            if x_step == 0 and y_step == 0:
                continue

            # Calcula os pr√≥ximos valores de x e de y
            next_x = self.x_values[-1] + x_step
            next_y = self.y_values[-1] + y_step

            self.x_values.append(next_x)
            self.y_values.append(next_y)

```

```python
while True:
    # Cria um passeio aleat√≥rio
    rw = RandomWalk(100000)
    rw.fill_walk()

    # Define o tamanho da janela de plotagem
    plt.figure(figsize=(10,6))

    # Plota e colore o passeio
    point_numbers = list(range(rw.num_points))
    plt.scatter(rw.x_values,rw.y_values,c=point_numbers,cmap=plt.cm.winter,s=1)
    
    # Enfatiza o primeiro e ultimo ponto
    plt.scatter(0,0,c='green',s=100)
    plt.scatter(rw.x_values[-1],rw.y_values[-1],c='red',s=100)

    # Remove os eixos
    plt.gca().get_xaxis().set_visible(False)
    plt.gca().get_yaxis().set_visible(False)

    plt.show()
    
    keep_running = input('Outro passeio? (y/n): ')
    if keep_running == 'n':
        break
```

```python
class RandomWalk2():
    """Classe para gerar passeios aleat√≥rios"""
    def __init__(self,num_points=5000):
        """Inicializa os atributos de um passeio"""
        self.num_points = num_points

        # Todos os passeio come√ßam em (0,0)
        self.x_values = [0]
        self.y_values = [0]

    def get_step(self):
        # Decide dire√ß√£o a ser seguida e dist√¢ncia a ser percorrida nela
            direction = choice([1,-1])
            distance = choice([0,1,2,3,5,8,13,21,34])
            return direction * distance

    def fill_walk(self):
        """Calcula todos os pontos do passeio"""
        # Continua dando passos at√© que o passeio alcance o tamanho desejado
        while len(self.x_values) < self.num_points:
            # Decide dire√ß√£o a ser seguida e dist√¢ncia a ser percorrida nela
            x_step = self.get_step()
            y_step = self.get_step()

            # Rejeita movimentos que n√£o v√£o a lugar nenhum
            if x_step == 0 and y_step == 0:
                continue

            # Calcula os pr√≥ximos valores de x e de y
            next_x = self.x_values[-1] + x_step
            next_y = self.y_values[-1] + y_step

            self.x_values.append(next_x)
            self.y_values.append(next_y)

# Cria um passeio aleat√≥rio
rw = RandomWalk2(50000)
rw.fill_walk()

# Define o tamanho da janela de plotagem
plt.figure(figsize=(10,6))

# Plota e colore o passeio
point_numbers = list(range(rw.num_points))
plt.plot(rw.x_values,rw.y_values,linewidth=1)

# Enfatiza o primeiro e ultimo ponto
plt.scatter(0,0,c='green',s=100)
plt.scatter(rw.x_values[-1],rw.y_values[-1],c='red',s=100)

# Remove os eixos
plt.gca().get_xaxis().set_visible(False)
plt.gca().get_yaxis().set_visible(False)

plt.show()
```

```python
from random import randint

class Die():
    """Uma classe que representa um √∫nico dado"""
    def __init__(self,num_sides=6):
        """Sup√µe que seja um dado de seis lados"""
        self.num_sides = num_sides

    def roll(self):
        """Devolve um valor aleat√≥rio entre 1 e o n√∫mero de lados"""
        return randint(1,self.num_sides)

# Cria um D6
die = Die()

# Faz alguns lan√ßamentos e armazena os resultados em uma lista
results = []
for roll_num in range(1000):
    result = die.roll()
    results.append(result)

# Analise resultados
frequencies =[]
for value in range(1,die.num_sides + 1):
    frequency = results.count(value)
    frequencies.append(frequency)

import pygal
# Visualiza os resultados
# Define o tipo de gr√°fico
hist = pygal.Bar()

# Formata o gr√°fico no pygal
hist.title = 'Resultados do dado'
hist.x_labels = ['1','2','3','4','5','6']
hist.x_title = 'Resultados'
hist.y_title = 'Frequ√™ncia de resultado'

# Adiciona legenda
hist.add('D6',frequencies)

# Exporta o gr√°fico
hist.render_to_file('die_visual.svg')
```
